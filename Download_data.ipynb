```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
TICKERS = ["MSFT", "KO", "TSLA", "AMD", "MS", "TSM", "AAPL"]
START = "2020-01-01"
END = "2025-01-01"
```

```python
data = yf.download(TICKERS, start=START, end=END)
data = data.drop(columns=["Volume"])
```

```python
data.describe()
```

```python
log_returns = np.log(data["Close"] / data["Close"].shift(1))
```

```python
# Basic Graph to for visual check
fig, ax = plt.subplots()
ax.plot(log_returns.cumsum())
fig.legend(log_returns.columns)
```

```python
data.to_csv("data.csv")
```
